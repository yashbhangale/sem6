```python
# Step 1: Loading time series dataset correctly in Pandas
import pandas as pd
from matplotlib import pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
data = pd.read_csv('Electric_Production.csv')  # Replace 'Electric_Production.csv' with the path to your dataset
data['DATE'] = pd.to_datetime(data['DATE'])
data.set_index('DATE', inplace=True)

# Step 3: Time-Resampling using Pandas
# Resample the data to monthly frequency ('M' for month end frequency)
monthly_data = data.resample('M').mean()

# Step 4: Rolling Time Series
# Calculate the rolling mean with a window of 3 months
rolling_mean = data.rolling(window=3).mean()

# Step 5: Plotting Time-series Data using Pandas
# Plot original data
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['IPG2211A2N'], label='Original')
plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()

# Plot rolling mean
plt.figure(figsize=(10, 6))
plt.plot(rolling_mean.index, rolling_mean['IPG2211A2N'], label='Rolling Mean')
plt.title('Rolling Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Rolling Mean')
plt.legend()
plt.show()

# Step 6: Bar Plot
# Plotting the resampled data as a bar plot
monthly_data.plot(kind='bar', figsize=(10, 6))
plt.title('Monthly Electric Production')
plt.xlabel('Date')
plt.ylabel('Production')
plt.show()

# Step 7: Autocorrelation Plot
# Plotting autocorrelation function (ACF)
plot_acf(data['IPG2211A2N'], lags=30)  # Adjust the number of lags as needed
plt.title('Autocorrelation Plot')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()

# Step 8: Seasonal Decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(data["IPG2211A2N"], model='additive')
result.plot()
plt.show()
```
